```python
import os
# 필요한 라이브러리 임포트
from langchain_upstage import UpstageEmbeddings
from dotenv import load_dotenv
from langchain_chroma import Chroma
import mysql.connector

load_dotenv()

# MySQL 데이터베이스 연결 설정
db_connection = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="0000",
    database="dbaas_be"
)

# 커서 생성
cursor = db_connection.cursor()

# 데이터베이스의 모든 테이블 이름 가져오기
cursor.execute("SHOW TABLES")
tables = [table[0] for table in cursor.fetchall()]

all_docs = []

# 각 테이블에 대해 데이터 가져오기
for table_name in tables:
    cursor.execute(f"SELECT * FROM {table_name}")
    results = cursor.fetchall()
    
    # 데이터 전처리 (필요한 경우)
    # docs = [str(row) for row in results]
    docs = [", ".join(map(str, row)) for row in results]
    all_docs.extend(docs)

print(results)
print("==========================")
print(all_docs)

embeddings = UpstageEmbeddings(model='solar-embedding-1-large-query')

index_name = "dbaas-index"

# 문서 임베딩 및 저장
Chroma.from_texts(texts=all_docs, 
                             embedding=embeddings, 
                             collection_name='chroma-dbaas', 
                             persist_directory="./chroma")

# 데이터베이스 연결 종료
cursor.close()
db_connection.close()
```
